```python
#Problem - 7 Find the possible key length second step.

from collections import defaultdict
from math import gcd

def find_common_factors(numbers):
    factors_count = defaultdict(int)
    for num in numbers:
        for i in range(2, num + 1): 
            while num % i == 0:
                factors_count[i] += 1
                num //= i

    return factors_count

all_distances = [distance for distances in distances.values() for distance in distances]

common_factors = find_common_factors(all_distances)

sorted_factors = sorted(common_factors.items(), key=lambda x: x[1], reverse=True)

print(sorted_factors)
```
